```python
data = data.dropna(subset=['floors_total'])
data['floors_total'] = data['floors_total'].astype('int')
```

```python
for column in ['balcony', 'parks_around3000', 'ponds_around3000']:
    data[column] = data[column].fillna(0)
```

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d %H:%M')
```

```python
data['locality_name'] = data['locality_name'].str.replace('ё','е', regex=True)
# list(data['locality_name'].sort_values().unique())
duplicates = ['городской поселок ', 'деревня ', 'коттеджный поселок ', 'поселок городского типа ', 'поселок станции ', 'поселок ', 'поселок при железнодорожной станции ', 'садовое товарищество ', 'село ', 'при железнодорожной ', 'садоводческое некоммерческое товарищество ']
name = ''
data['locality_name'] = data['locality_name'].replace(duplicates, name, regex=True)
# list(data['locality_name'].sort_values().unique())
data = data.dropna(subset=['locality_name'])
```

```python
data['ceiling_height'].where(~(data['ceiling_height'] >= 20) , other=data['ceiling_height']/10, inplace=True)
# list(data['ceiling_height'].sort_values().unique())
```

```python
data = data.rename(columns={'cityCenters_nearest' : 'city_centers_nearest'})

data['sq_m_cost'] = data['last_price'] / data['total_area']
```

```python
data['weekdays'] = data['first_day_exposition'].dt.weekday

data['months'] = data['first_day_exposition'].dt.month

data['years'] = data['first_day_exposition'].dt.year
```

```python
def floor_type(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row['floors_total']:
        return 'последний'
    else:
        return 'другой'
```

```python
data['floor_type'] = data.apply(floor_type, axis=1)
```

```python
data['city_centers_nearest_km'] = data['city_centers_nearest'] / 1000
data['city_centers_nearest_km'] = data['city_centers_nearest_km'].fillna(50).astype('int')
```
